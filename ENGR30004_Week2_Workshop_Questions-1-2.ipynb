### II.1 Functions 

There are two main ways to write a function in Python:


1.   Using a Function Definition: Frequently used, and preferred in python programs. Allows complex function definitions. 
2.   Using Lambda Operators: Harder to read, but highly useful sometimes. Allows for definition of mathematical operations etc. to be done in one line!



````python
# This is a function definition
def add_two_numbers(a, b):
    sum = a + b
    return sum


# This is a lambda operator 
y = lambda a, b : a + b
# Although lambda is a concise way to write simple function, it should be note 
# that we do not suggest you use lambda this way, as regular function def is 
# much clearer for reading and simpler for debugging.

m = 10
n = 22

# print('{} is the output of the defined function'.format(add_two_numbers(m, n)))
# print('{} is the output of the lambda operator'.format(y(m, n)))

print(f"{  add_two_numbers(m, n)  } is the output of the defined funtion")
print(f"{  y(m, n)  } is the output of the lambda operator")
````

````python
# (need to run previous cell before this one)
# In python, function is an object and can be passed to other functions
def apply_func(func, a, b):
    return func(a, b)


print(f"{apply_func(add_two_numbers, m, n)}"
      " is the output of the defined function")

# we suggest you use lambda when you certainly need to pass function as objects 
# and it is absolutely correct to do so. 
print(f"{apply_func(lambda a, b : a - b, m, n)}"
      " is the output of the lambda operator")
````

### Defining Arguments for Recursion

````python
def reverse(S, start, stop):
    if start < stop - 1:
        S[start], S[stop-1] = S[stop-1], S[start]
        reverse(S, start+1, stop-1)
    else:
        print(S)
````

````python
reverse([1, 2, 3, 4, 5, 6, 7, 8, 9], 0, 9)
````

### Functions (Further explanations - Please explore in your own time)

````python
# when calling a function, we can ignore the order if we pass values with argument names (i.e., keys)
def weighted_add(a, b):
    return a * 2.0 + b / 2.0


print('Without keys:')
print('weighted_add(0.5, 2.0) ->', weighted_add(0.5, 2.0))
print('weighted_add(2.0, 0.5) ->', weighted_add(2.0, 0.5))
print('With keys:')
print('weighted_add(a=0.5, b=2.0) ->', weighted_add(a=0.5, b=2.0))  # this is the same as weighted_add(0.5, 2.0)
print('weighted_add(b=2.0, a=0.5) ->', weighted_add(b=2.0, a=0.5))  # this is the same as weighted_add(0.5, 2.0)
print('weighted_add(b=0.5, a=2.0) ->', weighted_add(b=0.5, a=2.0))
````

````python
"""
We can mix key-value paired arguments and those with values only. However, 
there are two rules:
1. key-value pairs should appear after those without keys;
2. those without keys shall be passed in the exact same order in the function definition;

"""
print('weighted_add(0.5, b=2.0) ->', weighted_add(0.5, b=2.0))  # this is the same as passing a=0.5, b=2.0
print('weighted_add(0.5, a=2.0) ->', weighted_add(0.5, a=2.0))  # this gives an error
````

````python
"""
We can pass an arbitrary number of value arguments and key-value paired 
arguments using *args and **kwargs.
"""

def sum_all(*args, **kwargs):
    sum_args = 0
    sum_kwargs = 0
    for a in args:  # args is a tuple
        sum_args += a
    for b in kwargs.values():  # kwargs is a dictionary, dict.values() returns all values in dict
        sum_kwargs += b
    print('Sum of args is {0}, sum of kwargs is {1}'.format(sum_args, sum_kwargs))


sum_all(1, 2, 3, 4, x1=5, x2=6)
sum_all(1, 2, 3, 4)
sum_all(x1=5, x2=6)
````

````python
"""
We can mix regular arguments with *args and **kwargs
"""

def sum_all(a, *args, b=1, **kwargs):
    sum_args = a
    sum_kwargs = b
    for a in args:  # args is a tuple
        sum_args += a
    for b in kwargs.values():  # kwargs is a dictionary, dict.values() returns all values in dict
        sum_kwargs += b
    print('Sum of args is {0}, sum of kwargs is {1}'.format(sum_args, sum_kwargs))


sum_all(1, 2, 3, 4, x1=5, x2=6)
sum_all(1, 2, 3, 4)
sum_all(1, x1=5, x2=6)  # must provide a here
````

### **Object Orientation: Encapsulation and Abstraction**

Functions are pretty handy at reusing 'functional' (moving) parts of codes. How about organizing functionality based on a more *semantic* structure? 

By organizing a bunch of functions and associated variables into a single framework, we obtain an object called class.

#### Introduction

````python
"""
Class is an object encapsulation/abstraction of 
1. a group of variables, and 
2. a group of functions (we call a function in a class a 'method').  

Class is a template for its instances. We can initialize an instance using: 
  instance = Class(initial parameters)
They are similar to apartment deisgn blueprint and your apartment. Class is the 
blueprint, based on which, many apartments can be constructed which, at 
initialization, has the same structure. An instance is your apartment, where you 
may buy furniture to make it different from your neighbor's apartment. 

"""
class Cat(object): 
    """
    Here the Cat class inherits from an object called 'object', which is the template 
    for all Python classes. Simply speaking, "inherits" means we build Cat based on 
    the 'object' object. Using the apartment example above, 'object' is the blueprint 
    of all blueprints.
    """
    # Cat Taxonomy
    # below 'Kindom', 'Phylum', 'Class', 'Order' are class variables, belonging to 
    # the Cat class. If we change Cat.Kindom, we change it for all instances of Cat.
    Kingdom = 'Animalia'
    Phylum = 'Chordata'
    Class = 'Mammalia'
    Order = 'Carnivora'
  
    """
    By default, any defined function is a instance method.
    The first argument of an instance method is required to be a reference to 
    the instance itself. We can then add as many arguments as we want, just like we 
    define a regular function.
    In tradition, we use 'self' to refer to the instance. We can use any other name 
    but it is not suggested because it will only make your code unreadable to others. 

    __init__ is a built-in method of Python class. It is used to initialize an 
    instance. It is called when we first pass arguments to the class and the 
    initialization process implicitly returns self, e.g.,
    my_cat = Cat('biling', 'domestic long hair', 1)
    """
    def __init__(self, pet_name, pet_breed, pet_age): 
        # below 'name', 'breed' and 'age' are instance variables, which is similar 
        # to regular variables, but belongs to the instance of Cat class. After we 
        # initialize the class to get an instance, we can change the variables and it
        # will not affect other instances. 
        self.name = pet_name
        self.breed = pet_breed
        self.age = pet_age
  
    def say_hi(self):
        print('Hi, my name is {}. I\'m a {} year old {}'.format(self.name, self.age, self.breed))


# class instantiation, here we get an instance of Cat called my_cat
# at this step, the __init__ method of Cat is called.
my_cat = Cat('Tommy', 'domestic long hair', 1) # my cat is a domestic long hair, who's called biling, and is one year old

print(my_cat) # shows you that my_cat is an object
print(type(my_cat))  # shows you its type is Cat, i.e., it's an instance of Cat class
````

We cannot directly use a class function. First, we must 'intialize an object'. Much like you don't just have a 'template' cat, but rather a 'specific' cat. 

````python
# call a instance method
# there is no need to pass 'self' here. It is implicitly passed. 
my_cat.say_hi()
````

On the other hand, your cat may be a different one.

````python
# class instantiation, here we get another instance of Cat called your_cat
# my_cat and your_cat share the same class Cat, but they are different instances
your_cat = Cat('Timmy', 'russian blue', 2)

my_cat.say_hi()
your_cat.say_hi()
````

````python
# change made at intance level will not affect other instance
my_cat.name = 'Simba'

my_cat.say_hi()
your_cat.say_hi()
````

#### Object Abstraction and Inheritance

While we saw that we can pile related functions and variables into an object, we can go a bit further in terms of organization. We can define a hierarchy of objects. 

````python
class Pet(object):
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed
    
    def say(self):
        pass  # Pet can't say if we don't know what Pet is!
    
    def isAnimal(self):
        return True

class Cat(Pet):  # 'Cat' extends 'Pet'
    
    def say(self):  # this overwrites the default say method of Pet instances
        print('meow! I\'m {}'.format(self.name)) 

class Dog(Pet):  # 'Dog' extends 'Pet'
    
    def say(self):
        print('woof! I\'m {}'.format(self.name)) 
````

````python
tiny_cat = Cat('tim', 'persian')
big_dog = Dog('eric', 'dane')

tiny_cat.say()
big_dog.say()
````

````python
class Fox(Pet):  # 'Fox' extends 'Pet'
    def __init__(self):  # we do not need to follow the __init__ of Pet
        pass
    
    def say(self):
        print('What does the fox say? Ring-ding-ding-ding-dingeringeding!')

class Duck(Pet):
    
    def __init__(self, name, breed=None, weight=1):  
        # Furthermore, we can customize the init process, still call Pet.__init__ 
        # and more instance level variables
        super(Duck, self).__init__(name, breed)
        self.weight = weight
        
    def say(self):
        print('quack! I\'m {}'.format(self.name)) 


fox = Fox()
duck = Duck('Cheese')


fox.say()
duck.say()
````

````python
print(fox.isAnimal())
````

Here, we *abstracted* the concept of dogs and cats into one *pet*, as in our context, they both have shared and similar attributes and functions. In the object hierarchy, 'pet' is more abstract than 'cat'. 'dog' and 'cat' are of the same level!

When we *extend* the concept of 'pet' to 'cat/dog', 'cat/dog' *inherits* both properties and functions of its parent class (in this case 'pet'). However, we're free to manipulate the values of the child as we wish. 

We defined an empty function in the 'Pet' class, but we redefined it in the children classes. This is called **function overriding**.

#### Additional Concepts: Multiple Inheritance (Please explore in your own time)

Now, we saw that we can clearly inherit properties, both attributes and functions, from a single parent. Python provides a really nice ability to inherit properties from multiple parents. 


````python
class Person(object):
    
    def __init__(self, name): # every person has a name
        self.name = name
        
    def speak(self): # every person can say their name
        print('Hi, my name is {}.'.format(self.name))

class Mother(Person):
    
    def lift_weights(self): # mothers can lift weights
        print('Here goes {} lifting some weights again!'.format(self.name))

class Father(Person):
    
    def run_fast(self): # fathers can run fast
        print('Here comes {} running really fast again!'.format(self.name))

class Child(Mother, Father):
  
  def do_both(self):
    self.run_fast() # child has inherited fast running from father
    self.lift_weights() # they have inherited weight lifting from mother

lil_kid = Child("Lil Kid") # We create a Child called 'lil_kid'. 
lil_kid.speak() #lil_kid inherited speaking from 'person'.
lil_kid.do_both() 
````
